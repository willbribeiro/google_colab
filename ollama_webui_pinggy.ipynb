## check ollama running or not

```python
import subprocess
def start_ollama_server():
  subprocess.Popen(['ollama', 'serve'])
  print("Ollama server launched successfully!")
start_ollama_server()
```

## check the port where ollama is running

```python
def check_ollama_port(port='11434'):
  try:
    output = subprocess.run(['sudo', 'lsof', '-i', '-P', '-n'], capture_output=True, text=True).stdout
    if f":{port} (LISTEN)" in output:
      print(f"Ollama is actively listening on port {port}")
    else:
      print(f"Ollama is not detected on port {port}")
  except Exception as e:
    print(f"An error occurred: {e}")
check_ollama_port()
```
