# XTTS
XTTS é um modelo de geração de voz que permite clonar vozes em diferentes idiomas usando apenas um pequeno clipe de áudio de 6 segundos. Não há necessidade de uma quantidade excessiva de dados de treinamento que leve incontáveis ​​horas.

## Características
* Suporta 17 idiomas.
* Clonagem de voz com apenas um clipe de áudio de 6 segundos.
* Transferência de emoções e estilo por meio da clonagem.
* Clonagem de voz entre idiomas.
* Geração de fala multilíngue.
* Taxa de amostragem de 24 kHz.

## Línguas
O XTTS-v2 suporta 17 idiomas: inglês (en), espanhol (es), francês (fr), alemão (de), italiano (it), português (pt), polonês (pl), turco (tr), russo (ru), holandês (nl), tcheco (cs), árabe (ar), chinês (zh-cn), japonês (ja), húngaro (hu), coreano (ko) e hindi (hi).
