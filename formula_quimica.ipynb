```python
from rdkit import Chem, DataStructs
from rdkit.Chem import MolFromSmiles, Draw, Descriptors, AllChem
from rdkit.Chem.Draw import IPythonConsole
import PIL
from PIL import Image
import numpy as np
```

```python
smiles = 'CCCCCCCCCCC1=CC=C(C=C1)S(=O)(=O)[O-].[Na+]'
SDBS = Chem.MolFromSmiles(smiles)
SDBS
```

```python
MM = Descriptors.MolWt(SDBS)
print(f'A massa molar do SDBS é {MM:.2f} g/mol')
```
